```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
from cds import plotting

drive = pd.read_csv('Limma/lim_res_rnai_drive_NEW.csv',index_col=0)
rnai = pd.read_csv('Limma/lim_res_rnai_achilles_NEW.csv',index_col=0)
#crispr = pd.read_csv('Limma/lim_res_crispr.csv',index_col=0)
```

```python
# rnai_achilles -> rnai_drive at .25

sh = set(rnai.index & drive.index)



rnai_point_25 = rnai.loc[sh][rnai.loc[sh,'q.value'] < .25]
drive_point_25 = drive.loc[sh][drive.loc[sh,'q.value'] < .25]



pos_effect_size = rnai_point_25[rnai_point_25.loc[:,'EffectSize']<0]
top_100 = pos_effect_size.sort_values('EffectSize').iloc[0:100]
overlap = len(set(top_100.index & drive_point_25[drive_point_25.loc[:,'EffectSize']<0].index))

length_for_null = len(drive_point_25[drive_point_25.loc[:,'EffectSize']<0])
null = []
for x in range(0,1000000):
    randos = np.random.choice(drive.index,length_for_null,replace=False)
    null.append(len(set(randos) & set(top_100.index)))
    
fix,ax=plt.subplots(figsize=(6,4))
sns.distplot(null,kde_kws={'bw':1},label='null',bins=np.arange(np.array(null).min(), np.array(null).max()+1),
            hist_kws=dict(edgecolor="black", linewidth=1))
plt.axvline(sorted(null)[int(len(null) * .95)],ymax=.5,color='black',label='p=.05')
actual_p = ((sum(np.array(null) > overlap) +1) * 1.0) / (len(null)+1)
plt.axvline(overlap,color='red',ymax=.25,label='overlap\n(p=' + str(round(actual_p,7)) + ')')
plotting.simpleaxis(ax)
plt.ylabel('density')
plt.legend()
plt.xlabel('number of overlapping genes from RNAi Drive')
plt.title('RNAi Achilles / RNAi Drive ')
plt.savefig('fig1d/rnai_achilles_rnai_drive_overlap_25_0.pdf')
```

```python
from scipy import stats
sh = set(rnai.index&drive.index)


rnai_ = rnai.loc[sh][rnai.loc[sh,'EffectSize'] < 0]
drive_ = drive.loc[sh][drive.loc[sh,'EffectSize'] < 0]



thresh = .25
_100 = rnai_[rnai_.loc[:,'q.value'] < thresh].sort_values('EffectSize').iloc[0:100]

dataset_b = drive_[drive_.loc[:,'q.value'] < thresh]

both = set(set(_100.index) & set(dataset_b.index))

v=venn2(subsets = (len(set(_100.index) - both), len(set(dataset_b.index) - both), len(set(set(_100.index) & set(dataset_b.index)))), set_labels = ('RNAi Achilles','RNAi Drive'))



a_sig = set(_100.index)
b_sig = set(dataset_b.index)
a_not_sig = set(sh) - a_sig
b_not_sig = set(sh) - b_sig

oddsratio, pvalue = stats.fisher_exact([[len(a_sig & b_sig), len(a_sig & b_not_sig)],
                                        [len(a_not_sig & b_sig), len(a_not_sig & b_not_sig)]])

print(pvalue)

plt.legend(title='p = '+ str(round(pvalue,10)) + '\nat q-value of .25',frameon=False,loc='upper left')

plt.savefig('fig1d/achilles_drive_100_.25_0.pdf')
```
